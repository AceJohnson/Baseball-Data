```python
>>> from pybaseball import statcast
from pybaseball import playerid_lookup
from pybaseball import statcast_pitcher
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
playerid_lookup('degrom','jacob')
```

```python
degrom_stats = statcast_pitcher('2019-04-01', '2019-05-01', 594798)
```

```python
degrom_stats.head(10)
```

```python
degrom_spinrate = degrom_stats[['pitch_type','release_spin_rate','release_extension','description','events']]
```

```python
degrom_spinrate.head(10)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.despine(fig, left=True, bottom=True)
sns.scatterplot(x="release_spin_rate", y="release_extension",
                hue="events", 
                palette='viridis', 
                data=degrom_spinrate, ax=ax)
ax.set_title("Pitch Result by Spin Rate and Release");
```

```python
degrom_spinrate['strike'] = degrom_spinrate['description'].str.contains('strike')
```

```python
degrom_spinrate['hr'] = degrom_spinrate['events'] =='home_run'
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.despine(fig, left=True, bottom=True)
sns.scatterplot(x="release_spin_rate", y="release_extension",
                hue="hr", 
                palette='binary', 
                data=degrom_spinrate, ax=ax)
ax.set_title("Pitch Result by Spin Rate and Release");
```
